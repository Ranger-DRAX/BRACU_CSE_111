```python
#task-1
class Student:
    id=0
    def __init__(self,name,dept,age,cgpa):
        self.name=name
        self.dept=dept
        self.age=age
        self.cgpa=cgpa
        Student.id+=1
    def get_details(self):
        print(Student.id)
        print("Name:",self.name)
        print('Department:',self.dept)
        print('Age:',self.age)
        print("CGPA:",self.cgpa)
    @classmethod
    def from_String(cls,txt):
        a=txt.split('-')
        return Student(a[0],a[1],a[2],a[3])
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()
print('Subtask_5:')
print('class_variable:share among all objects,affects them all immediately')
print('instance_variable:Only affects the object saparetly')
print('Subtask_6:')
print('Instance_method:Instance methods which  can access the attributes and the other methods through the “self” parameter')
print('Class_methods:for calling classmethods we dont need to create any specific object')
```

```python
#task_2
class Assassin:
    t_n_a=0
    def __init__(self,name,rate):
        self.name=name
        self.rate=rate
        Assassin.t_n_a+=1
    def printDetails(self):
        print('Name:',self.name)
        print(f"Sucess rate:{self.rate}%")
        print('Total number of assassin:',Assassin.t_n_a)
    @classmethod
    def failureRate(cls,n,r):
        return Assassin(n,100-int(r))
    @classmethod
    def failurePercentage(cls,n1,r1):
        k = r1.split('%')
        a = int(k[0])
        return Assassin(n1,100-a)
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
```

```python
#task-5
class Employee:
    def __init__(self,n,h):
        self.name=n
        self.workingPeriod=h
    @classmethod
    def employeeByJoiningYear(cls,nam,year):
        a=str(year)[3]
        b=int(a)
        return Employee(nam,b)
    @staticmethod
    def experienceCheck(age,gender):
        if age>=3 :
            if gender=='male':
                return 'He is experienced'
            else:
              return 'She is experienced'
        else:
            if gender=='male':
              return 'He is not experienced'
            else:
              return 'She is not experienced'
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#task_7
class Cat:
    Number_of_cats=0
    def __init__(self,color,position):
        self.color=color
        self.position=position
        Cat.Number_of_cats+=1

    @classmethod
    def no_parameter(self):
        return Cat('White','sitting')

    @classmethod
    def first_parameter(cls,kk):
        return Cat(kk,'sitting')

    @classmethod
    def second_parameter(cls,gg):
        return Cat('Grey',gg)
    def changeColor(self,c1):
        self.color=c1
    def printCat(self):
        print(f"{self.color} is {self.position}")

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task_8
import math
class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

        print(f'Default radius={Cylinder.radius} and height={Cylinder.height}.')
        Cylinder.height = self.height
        Cylinder.radius = self.radius
        print(f'Updated: radius={Cylinder.radius} and height={Cylinder.height}.')

    @classmethod
    def swap(cls, height, radius):
        return Cylinder(radius, height)

    @classmethod
    def changeFormat(cls, string):
        info = string.split('-')
        return Cylinder(float(info[0]), float(info[1]))
    @staticmethod
    def area(radius, height):
        area = (2 * math.pi * (radius ** 2)) + (2 * math.pi * radius * height)
        print('Area:', area)
    @staticmethod
    def volume(radius, height):
        volume = math.pi * (radius ** 2) * height
        print('Volume:', volume)
c1 = Cylinder(0, 0)
Cylinder.area(c1.radius, c1.height)
Cylinder.volume(c1.radius, c1.height)
print("===============================")
c2 = Cylinder.swap(8, 3)
c2.area(c2.radius, c2.height)
c2.volume(c2.radius, c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius, c3.height)
c3.volume(c3.radius, c3.height)
print("===============================")
Cylinder(0.3, 5.56).area(Cylinder.radius, Cylinder.height)
print("===============================")
Cylinder(3, 5).volume(Cylinder.radius, Cylinder.height)
```

```python
#task_9
class Student:
    total_students=0
    bracu_students=0
    other_students=0
    def __init__(self,name,dept,institute='BRAC University'):
        self.name=name
        self.dept=dept
        self.ins=institute
        Student.total_students+=1
        if self.ins=="BRAC University":Student.bracu_students+=1
        else:Student.other_students+=1

    @classmethod
    def printDetails(cls):
        print(f"Total students:{Student.total_students}")
        print(f"BRAC university Students:{Student.bracu_students}")
        print(f"Other institution students:{Student.other_students}")
    @classmethod
    def createStudent(cls,n,d,u=None):
        if u!=None:
            b=u
            return Student(n,d,b)
        else:return Student(n,d)
    def individualDetail(self):
        print(f"Name:{self.name}")
        print(f"Department:{self.dept}")
        print(f"Institution:{self.ins}")

Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```
