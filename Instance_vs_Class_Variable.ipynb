```python
class BdCricketTeam:

  team_run = 0  # Class variable; Shared among all objects, affects them all immadiately.

  def __init__(self, player_name):
    self.player_run = 0   # instance variable; Only affects the object separtely
    self.player_name = player_name  # instance variable; Only affects the object separtely
    # self.team_run = 0

  def hit_four(self):
    self.player_run += 4
    BdCricketTeam.team_run += 4

  def hit_six(self):
    self.player_run += 6
    type(self).team_run += 6

  def took_run(self, run=1):
    self.player_run += run
    type(self).team_run += run

  def show_score(self):
    print(f"{self.player_name} scored : {self.player_run}")
    print(f"Total Team Score : {self.team_run}")
    print("==============================================")

Tamim = BdCricketTeam("Tamim")
Tamim.hit_four()
Tamim.took_run(2)
Tamim.hit_six()

Tamim.show_score()

Mash = BdCricketTeam("Mashrafi")
Mash.hit_four()
Mash.hit_six()

Mash.show_score()

Soummo  = BdCricketTeam("Soummo")
Soummo.took_run(2)
Soummo.hit_six()

Soummo.show_score()
```

```python
class BdCT:

  team_run = 0  # CLASS VARIABLE; shared among all objects
  player_count = 0 # CLASS VARIABLE; shared among all objects
  players = {}  # CLASS VARIABLE; shared among all objects ; key == player name & value =run

  def __init__(self, name):
    self.player_name = name # INSTANCE VARIABLE; affects only specific object
    self.player_run = 0   # INSTANCE VARIABLE; affects only specific object
    type(self).player_count += 1
    type(self).players[self.player_name] = self.player_run

    print(f"{self.player_name} entered the pitch")
    

  def hit_four(self):
    type(self).players[self.player_name] += 4
    type(self).team_run += 4

  def hit_six(self):
    type(self).players[self.player_name] += 6
    type(self).team_run += 6

  def took_run(self, run=1):
    type(self).players[self.player_name] += run
    type(self).team_run += run

  def show_score(self):
    print(f"{type(self).player_count} player(s) played till now : ")
    
    for player, run in type(self).players.items():
      print(f"{player} scored {run} run(s)")
  
    print(f"\nTotal Team Score : {self.team_run}")
    print("==============================================")


```

```python
Tamim = BdCT("Tamim")
Tamim.hit_four()
Tamim.hit_six()
Tamim.show_score()

Mash = BdCT("Mashrafi")
Mash.hit_four()
Mash.took_run(2)
Mash.show_score()


Soummo = BdCT("Soummo")
Soummo.hit_six()
Soummo.took_run(3)
Soummo.show_score()
```
