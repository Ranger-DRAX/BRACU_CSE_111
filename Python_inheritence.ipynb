# **Inheritance**

### **Single Inheritance**image.png

```python
# single inheritance example
 
class Parent:                  # parent class
  def func1(self):                   
    print(f"In Parent Class")
 
class Child(Parent):           # child class
  def func2(self):                 # we include the parent class
    print(f"In Child Class")   # as an argument in the child
                               # class
 
# Driver Code
c = Child()                 # object created
c.func1()                   # parent method called via child object
c.func2()                   # child method called
```

### **Multi-Level Inheritance**
image.png

```python
# multi-level inheritance example
class Grandparent:                 # first level
  disease = "Diabetes"
  def func1(self):                   
    print(f"Inside Grandparent")
 
class Parent(Grandparent):         # second level
  def func2(self):                   
    print(f"Inside Parent")
 
class Child(Parent):               # third level
  def func3(self):                   
    print(f"Inside Child")   
                               
 
# Driver Code
c = Child()                     # object created
c.func1()                       # 3rd level calls 1st level
c.func2()                       # 3rd level calls 2nd level
c.func3()                       # 3rd level calls 3rd level
print(c.disease)
print(help(Child))
```

### **Multiple Inheritance**
image.png

```python
#multiple inheritance example
 
class Parent1:                     # first parent class
    def func1(self):                   
        print(f"Inside Parent1")
 
class Parent2:                     # second parent class
    def func2(self):                   
        print(f"Inside Parent2")

class Child(Parent1, Parent2):     # child class
    def func3(self):                     # we include the parent classes
        print(f"Inside child")       # as an argument comma separated
                           
# Driver Code
c = Child()        # object created
c.func1()          # parent1 method called via child
c.func2()          # parent2 method called via child instead of parent3
c.func3()          # child method called
 
# to find the order of classes visited by the child class, we use __mro__ on the child class
print(Child.__mro__)
print(help(Child))
```

### **Hybrid Inheritance**

```python
#ybrid inheritance example
 
class Parent1:                            # first parent class
    def func1(self):                   
        print(f"Inside Parent1")
 
 
class Parent2:                            # second parent class
    def func2(self):                   
        print(f"Inside Parent2")
 
class Child1(Parent1):                    # first child class
    def func3(self):                   
        print(f"Inside Child1")
 
 
class Child2(Child1, Parent2):            # second child class
    def func4(self):                   
        print(f"Inside Child2")   
                               
 
# Driver Code
c1 = Child1()                          # object created
c2 = Child2()
print("======================================================")

c1.func1()                       # child1 calling parent1 method
c1.func3()                       # child1 calling its own method
print("======================================================")

c2.func1()                       # child2 calling parent1 method
c2.func2()                       # child2 calling parent2 method
c2.func3()                       # child2 calling child1 method
c2.func4()                       # child2 calling its own method
print("======================================================")
```

## **Example of Variable and Method Overriding**

```python
class GrandParent:
  _title = "Khan"
  __wealth = "1 Crore"
  def func1(self):
    print("Inside GrandParent Class")
  def func4(self):
    print("Inside GrandParent Class 2")

class Parent(GrandParent):
  _title = "Ullah" # Variable Overriding
  def func2(self):
    print("Inside Parent Class")

class Child(Parent):
  def func1(self):
    print("Party !!!")  # Method Overriding
  def func3(self):
    print("Inside Child Class")
```

```python
c2 = Child()
c2.func1()
c2.func2()
c2.func3()

print(c2._title)

# print(c2.__wealth) # Error. Because Private Variable doesn't pass through the descendants 

help(c2)
```

### **The use of `super()` method**

*   `super()` points to the ***immediate*** **parent** class
*   used to ***call the method*** of immedaite parent class


```python
class Parent:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.gender = "Male"
        self.hair_color = "Black"

class Child(Parent):
    def __init__(self, name, age, technology):
        super().__init__(name, age)
        self.technology = technology

    def print_details(self):
        print(f"Name : {self.name}")
        print(f"Age: {self.age}")
        print(f"Technology : {self.technology}")

class Grandchild(Child):
    def __init__(self, name, age, tech, pet):
        super().__init__(name, age, technology=tech)
        self.pet = pet

c1 = Child("John", 30, "Python")
c1.print_details()

gc1 = Grandchild("Susan", 6, "Rust", "Nanobot")
gc1.print_details()
```

### **What if we want to call / use method `parent to parent (Grandparent) class`' method?**

```python
class Father:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def print_details(self):
        print("Inside Father class")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

f1 = Father("John", 45)
f1.print_details()


class Son(Father):
    def __init__(self, s_name, s_age, tech):
        super().__init__(s_name, s_age)
        self.tech = tech

    def print_details(self):
        print("Inside Son class")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Tech: {self.tech}")

s1 = Son("Jasson", 30, "Python")
s1.print_details()

class Grandson(Son):
    def __init__(self, name, age, tech, pet):
        Father.__init__(self, name, age)
        self.tech = tech
        self.pet = pet
            
    def print_details(self):
        print("Inside Grandson class")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        # print(f"Tech: {self.tech}")
        print(f"Pet : {self.pet}")
    

g1 = Grandson("Jonnas", 6, "Java", "Nanobot")
g1.print_details()
```

```python
print(isinstance(f1, Son)) #  is f1 an instance  (object) of Son class ?
print(issubclass(Grandson, Father)) # is Grandson a subclass of Father class?
```
